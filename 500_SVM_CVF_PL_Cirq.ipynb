```python
# ============================================
# SVM BASELINE - 500 samples, 7 runs
# Fast, simple, gives you mean ± std for paper
# ============================================

import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import time
import pandas as pd
from datetime import datetime

print("="*60)
print("🚀 SVM BASELINE: 500 samples, 7 runs")
print("="*60)

# Load MNIST
print("\n📦 Loading MNIST (0 vs 1)...")
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target.astype(int)

# Filter 0 and 1
mask = (y == 0) | (y == 1)
X, y = X[mask], y[mask]
print(f"Total 0/1 samples: {len(y)}")

# Scale once
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store results
svm_results = []

for run in range(7):
    print(f"\n🔷 Run {run+1}/7 (seed={42+run})")

    # Different split each run
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.3, random_state=42+run, stratify=y
    )

    # PCA to 2D
    pca = PCA(n_components=2)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # TAKE EXACTLY 500 SAMPLES
    X_train_500 = X_train_pca[:500]
    y_train_500 = y_train[:500]
    X_test_250 = X_test_pca[:250]  # Proportional test set
    y_test_250 = y_test[:250]

    print(f"   Train: {len(X_train_500)} samples, Test: {len(X_test_250)} samples")

    # Train SVM
    start = time.time()
    svm = SVC(kernel='rbf', C=1.0, probability=True)
    svm.fit(X_train_500, y_train_500)
    y_pred = svm.predict(X_test_250)
    y_prob = svm.predict_proba(X_test_250)[:, 1]

    # Calculate ALL metrics
    from sklearn.metrics import roc_auc_score
    svm_results.append({
        'run': run+1,
        'accuracy': accuracy_score(y_test_250, y_pred) * 100,
        'precision': precision_score(y_test_250, y_pred, average='binary'),
        'recall': recall_score(y_test_250, y_pred, average='binary'),
        'f1': f1_score(y_test_250, y_pred, average='binary'),
        'roc_auc': roc_auc_score(y_test_250, y_prob),
        'time': time.time() - start
    })

    print(f"   ✅ Acc={svm_results[-1]['accuracy']:.1f}%, F1={svm_results[-1]['f1']:.3f}, Time={svm_results[-1]['time']:.2f}s")

# ============================================
# RESULTS WITH STATISTICS
# ============================================
print("\n" + "="*60)
print("📊 SVM RESULTS: 7 runs with 500 samples")
print("="*60)

df = pd.DataFrame(svm_results)

print(f"\n{'Metric':<15} {'Mean':<10} {'Std':<10} {'95% CI':<20}")
print("-" * 55)

from scipy import stats

for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc', 'time']:
    values = df[metric].values
    mean_val = np.mean(values)
    std_val = np.std(values)
    ci = stats.t.interval(0.95, len(values)-1, loc=mean_val, scale=stats.sem(values))

    if metric == 'accuracy':
        print(f"{metric.capitalize():<15} {mean_val:.1f}%    ±{std_val:.1f}%    [{ci[0]:.1f}%, {ci[1]:.1f}%]")
    elif metric == 'time':
        print(f"{metric.capitalize():<15} {mean_val:.2f}s   ±{std_val:.2f}s   [{ci[0]:.2f}s, {ci[1]:.2f}s]")
    else:
        print(f"{metric.capitalize():<15} {mean_val:.3f}    ±{std_val:.3f}    [{ci[0]:.3f}, {ci[1]:.3f}]")

# ============================================
# SAVE RESULTS
# ============================================
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"svm_500samples_7runs_{timestamp}.csv"
df.to_csv(filename, index=False)
print(f"\n💾 Saved to {filename}")

# ============================================
# QUICK PLOT
# ============================================
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Accuracy across runs
ax1.plot(range(1, 8), df['accuracy'], 'bo-', linewidth=2, markersize=8)
ax1.axhline(y=df['accuracy'].mean(), color='r', linestyle='--', label=f"Mean: {df['accuracy'].mean():.1f}%")
ax1.fill_between(range(1, 8),
                 df['accuracy'].mean() - df['accuracy'].std(),
                 df['accuracy'].mean() + df['accuracy'].std(),
                 alpha=0.2, color='r')
ax1.set_xlabel('Run')
ax1.set_ylabel('Accuracy (%)')
ax1.set_title('SVM Accuracy Across 7 Runs (500 samples)')
ax1.set_ylim(95, 100)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Metrics bar chart
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC']
values = [df['accuracy'].mean()/100, df['precision'].mean(), df['recall'].mean(), df['f1'].mean(), df['roc_auc'].mean()]
errors = [df['accuracy'].std()/100, df['precision'].std(), df['recall'].std(), df['f1'].std(), df['roc_auc'].std()]

ax2.bar(metrics, values, yerr=errors, capsize=5, color='#2ecc71', edgecolor='black', alpha=0.7)
ax2.set_ylabel('Score')
ax2.set_title('SVM Metrics (mean ± std)')
ax2.set_ylim(0.95, 1.0)
ax2.grid(True, alpha=0.3, axis='y')

plt.suptitle('Classical SVM Baseline: 500 Samples, 7 Runs', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('svm_500samples_results.png', dpi=250)
plt.show()

print("\n✅ DONE! SVM baseline complete!")
```

##SVM Cross-Validation Basline

```python
# ============================================
# SVM CROSS-VALIDATION BASELINE
# Optimized for Google Colab
# ============================================

# 1. Colab Specific Setup
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')  # Suppress sklearn/openml warnings

# 2. Installations (Usually pre-installed, but safe to ensure)
# !pip install tqdm scikit-learn pandas matplotlib numpy scipy

# 3. Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from scipy import stats
from datetime import datetime
from tqdm import tqdm  # Progress bar

print("="*60)
print("🚀 SVM CROSS-VALIDATION BASELINE")
print("="*60)

# ============================================
# CONFIGURATION
# ============================================
# Set to True if you want to compare strictly with your previous
# "500 samples" experiment (will use 500 total samples for CV).
# Set to False for a robust baseline using ALL available 0/1 data.
LIMIT_DATA_TO_500 = False
N_SPLITS = 5  # 5-Fold Cross-Validation
RANDOM_STATE = 42

# ============================================
# LOAD & PREPARE DATA
# ============================================
print("\n📦 Loading MNIST (0 vs 1)...")
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target.astype(int)

# Filter 0 and 1
mask = (y == 0) | (y == 1)
X, y = X[mask], y[mask]
print(f"Total 0/1 samples available: {len(y)}")

# Optional: Restrict to 500 total samples for comparison
if LIMIT_DATA_TO_500:
    print(f"⚠️  LIMITING DATASET TO 500 TOTAL SAMPLES (Stratified)")
    from sklearn.model_selection import train_test_split
    # We use stratified sampling to get exactly 500 representative samples
    X, _, y, _ = train_test_split(X, y, train_size=500, stratify=y, random_state=RANDOM_STATE)
    print(f"   Dataset size reduced to: {len(y)}")

# Scale Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA (Optional, but kept from your original code for consistency)
# Note: For SVM on MNIST, PCA isn't strictly necessary, but keeps experiment consistent
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# ============================================
# CROSS-VALIDATION LOOP
# ============================================
print(f"\n🔷 Starting {N_SPLITS}-Fold Stratified Cross-Validation...")

kfold = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=RANDOM_STATE)
svm_results = []

# Use tqdm for a nice progress bar in Colab
for fold, (train_idx, test_idx) in enumerate(tqdm(kfold.split(X_pca, y), total=N_SPLITS)):

    # Split data using CV indices
    X_train, X_test = X_pca[train_idx], X_pca[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # Train SVM
    svm = SVC(kernel='rbf', C=1.0, probability=True, random_state=RANDOM_STATE)
    svm.fit(X_train, y_train)

    # Predict
    y_pred = svm.predict(X_test)
    y_prob = svm.predict_proba(X_test)[:, 1]

    # Calculate Metrics
    svm_results.append({
        'fold': fold + 1,
        'accuracy': accuracy_score(y_test, y_pred) * 100,
        'precision': precision_score(y_test, y_pred, average='binary'),
        'recall': recall_score(y_test, y_pred, average='binary'),
        'f1': f1_score(y_test, y_pred, average='binary'),
        'roc_auc': roc_auc_score(y_test, y_prob),
        'train_size': len(y_train),
        'test_size': len(y_test)
    })

# ============================================
# RESULTS WITH STATISTICS
# ============================================
print("\n" + "="*60)
print("📊 CV RESULTS SUMMARY")
print("="*60)

df = pd.DataFrame(svm_results)

print(f"\n{'Metric':<15} {'Mean':<10} {'Std':<10} {'95% CI':<20}")
print("-" * 55)

for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']:
    values = df[metric].values
    mean_val = np.mean(values)
    std_val = np.std(values)
    ci = stats.t.interval(0.95, len(values)-1, loc=mean_val, scale=stats.sem(values))

    if metric == 'accuracy':
        print(f"{metric.capitalize():<15} {mean_val:.1f}%    ±{std_val:.1f}%    [{ci[0]:.1f}%, {ci[1]:.1f}%]")
    else:
        print(f"{metric.capitalize():<15} {mean_val:.3f}    ±{std_val:.3f}    [{ci[0]:.3f}, {ci[1]:.3f}]")

# ============================================
# SAVE RESULTS
# ============================================
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
data_tag = "500samples" if LIMIT_DATA_TO_500 else "FullData"
filename = f"svm_cv_{data_tag}_{N_SPLITS}fold_{timestamp}.csv"
df.to_csv(filename, index=False)
print(f"\n💾 Saved to {filename}")

# ============================================
# PLOTTING
# ============================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 1. Accuracy across Folds
ax1.plot(range(1, N_SPLITS + 1), df['accuracy'], 'bo-', linewidth=2, markersize=8, label='Fold Accuracy')
ax1.axhline(y=df['accuracy'].mean(), color='r', linestyle='--', label=f"Mean: {df['accuracy'].mean():.1f}%")
ax1.fill_between(range(1, N_SPLITS + 1),
                 df['accuracy'].mean() - df['accuracy'].std(),
                 df['accuracy'].mean() + df['accuracy'].std(),
                 alpha=0.2, color='r', label='±1 Std Dev')
ax1.set_xlabel('Fold')
ax1.set_ylabel('Accuracy (%)')
ax1.set_title(f'SVM Accuracy Across {N_SPLITS} Folds\n({data_tag})')
ax1.set_ylim(90, 100) # Adjusted ylim for CV variance
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Metrics Bar Chart
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC']
# Normalize accuracy to 0-1 for the bar chart to match others
values = [df['accuracy'].mean()/100, df['precision'].mean(), df['recall'].mean(), df['f1'].mean(), df['roc_auc'].mean()]
errors = [df['accuracy'].std()/100, df['precision'].std(), df['recall'].std(), df['f1'].std(), df['roc_auc'].std()]

ax2.bar(metrics, values, yerr=errors, capsize=5, color='#3498db', edgecolor='black', alpha=0.7)
ax2.set_ylabel('Score')
ax2.set_title('SVM Metrics (Mean ± Std)')
ax2.set_ylim(0.90, 1.0)
ax2.grid(True, alpha=0.3, axis='y')

plt.suptitle(f'SVM Cross-Validation Baseline: {N_SPLITS}-Fold ({data_tag})', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'svm_cv_{data_tag}_results.png', dpi=250)
plt.show()

print("\n✅ DONE! Cross-Validation complete!")
```

```python
# ============================================
# CELL 1: PENNYLANE BASIC CIRCUIT (NO ZZ)
# H + RY encoding only - 500 samples, 7 runs
# ============================================

!pip install pennylane scikit-learn pandas --quiet

import numpy as np
import pennylane as qml
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import time
import pandas as pd
from scipy import stats
from datetime import datetime

print("="*60)
print("🔹 CELL 1: PENNYLANE BASIC (NO ZZ) - 500 samples, 7 runs")
print("="*60)

# Load MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target.astype(int)
mask = (y == 0) | (y == 1)
X, y = X[mask], y[mask]

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store results
results_basic = []

for run in range(7):
    print(f"\n🔷 Run {run+1}/7 (seed={42+run})")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.3, random_state=42+run, stratify=y
    )

    # PCA to 2D
    pca = PCA(n_components=2)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # TAKE EXACTLY 500 SAMPLES
    X_train_500 = X_train_pca[:500]
    y_train_500 = y_train[:500]
    X_test_250 = X_test_pca[:250]
    y_test_250 = y_test[:250]

    # Scale for quantum circuits
    X_train_scaled = X_train_500 * np.pi
    X_test_scaled = X_test_250 * np.pi

    print(f"   Train: {len(X_train_500)} samples, Test: {len(X_test_250)} samples")

    # BASIC CIRCUIT DEVICE
    dev_basic = qml.device("default.qubit", wires=2)

    @qml.qnode(dev_basic)
    def get_state_basic(x):
        """Basic H + RY encoding - NO ZZ"""
        qml.Hadamard(wires=0)
        qml.Hadamard(wires=1)
        qml.RY(x[0], wires=0)
        qml.RY(x[1], wires=1)
        return qml.state()

    start_basic = time.time()

    # Pre-compute states
    states_train_basic = [get_state_basic(x) for x in X_train_scaled]
    states_test_basic = [get_state_basic(x) for x in X_test_scaled]

    # Compute kernel matrix
    print("   Computing kernel matrices...")
    K_train_basic = np.zeros((500, 500))
    for i in range(500):
        for j in range(i, 500):
            val = qml.math.fidelity_statevector(states_train_basic[i], states_train_basic[j])
            K_train_basic[i,j] = val
            K_train_basic[j,i] = val

    K_test_basic = np.zeros((250, 500))
    for i in range(250):
        for j in range(500):
            K_test_basic[i,j] = qml.math.fidelity_statevector(states_test_basic[i], states_train_basic[j])

    # Train QSVM
    qsvm_basic = SVC(kernel='precomputed', probability=True)
    qsvm_basic.fit(K_train_basic, y_train_500)
    y_pred_basic = qsvm_basic.predict(K_test_basic)
    y_score_basic = qsvm_basic.decision_function(K_test_basic)
    y_score_norm_basic = (y_score_basic - y_score_basic.min()) / (y_score_basic.max() - y_score_basic.min() + 1e-10)

    time_basic = time.time() - start_basic

    # Store results
    results_basic.append({
        'run': run+1,
        'circuit': 'Basic H+RY (NO ZZ)',
        'accuracy': accuracy_score(y_test_250, y_pred_basic) * 100,
        'precision': precision_score(y_test_250, y_pred_basic, average='binary'),
        'recall': recall_score(y_test_250, y_pred_basic, average='binary'),
        'f1': f1_score(y_test_250, y_pred_basic, average='binary'),
        'roc_auc': roc_auc_score(y_test_250, y_score_norm_basic),
        'time': time_basic
    })

    print(f"   ✅ Acc={results_basic[-1]['accuracy']:.1f}%, F1={results_basic[-1]['f1']:.3f}, Time={time_basic:.1f}s")

# ============================================
# RESULTS
# ============================================
print("\n" + "="*60)
print("📊 BASIC CIRCUIT (NO ZZ) - RESULTS")
print("="*60)

df_basic = pd.DataFrame(results_basic)

print(f"\n{'Metric':<15} {'Mean':<10} {'Std':<10} {'95% CI':<20}")
print("-" * 55)

for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc', 'time']:
    values = df_basic[metric].values
    mean_val = np.mean(values)
    std_val = np.std(values)
    ci = stats.t.interval(0.95, len(values)-1, loc=mean_val, scale=stats.sem(values))

    if metric == 'accuracy':
        print(f"{metric.capitalize():<15} {mean_val:.1f}%    ±{std_val:.1f}%    [{ci[0]:.1f}%, {ci[1]:.1f}%]")
    elif metric == 'time':
        print(f"{metric.capitalize():<15} {mean_val:.1f}s   ±{std_val:.1f}s   [{ci[0]:.1f}s, {ci[1]:.1f}s]")
    else:
        print(f"{metric.capitalize():<15} {mean_val:.3f}    ±{std_val:.3f}    [{ci[0]:.3f}, {ci[1]:.3f}]")

# Save
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"pennylane_basic_NOzz_500_{timestamp}.csv"
df_basic.to_csv(filename, index=False)
print(f"\n💾 Saved to {filename}")
print("\n✅ CELL 1 COMPLETE! Run CELL 2 next.")
```

```python
# ============================================
# CELL 2: PENNYLANE ZZFEATUREMAP (WITH ZZ)
# H + RZ + CZ + RZ - 500 samples, 7 runs
# ============================================

!pip install pennylane scikit-learn pandas --quiet

import numpy as np
import pennylane as qml
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import time
import pandas as pd
from scipy import stats
from datetime import datetime

print("="*60)
print("🔸 CELL 2: PENNYLANE ZZFEATUREMAP (WITH ZZ) - 500 samples, 7 runs")
print("="*60)

# Load MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target.astype(int)
mask = (y == 0) | (y == 1)
X, y = X[mask], y[mask]

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store results
results_zz = []

for run in range(7):
    print(f"\n🔷 Run {run+1}/7 (seed={42+run})")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.3, random_state=42+run, stratify=y
    )

    # PCA to 2D
    pca = PCA(n_components=2)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # TAKE EXACTLY 500 SAMPLES
    X_train_500 = X_train_pca[:500]
    y_train_500 = y_train[:500]
    X_test_250 = X_test_pca[:250]
    y_test_250 = y_test[:250]

    # Scale for quantum circuits
    X_train_scaled = X_train_500 * np.pi
    X_test_scaled = X_test_250 * np.pi

    print(f"   Train: {len(X_train_500)} samples, Test: {len(X_test_250)} samples")

    # ZZFEATUREMAP DEVICE
    dev_zz = qml.device("default.qubit", wires=2)

    @qml.qnode(dev_zz)
    def get_state_zz(x):
        """ZZFeatureMap with 2 repetitions"""
        # Initial Hadamards
        for i in range(2):
            qml.Hadamard(wires=i)

        # 2 repetitions of ZZFeatureMap
        for _ in range(2):
            # RZ rotations
            for i in range(2):
                qml.RZ(x[i], wires=i)
            # CZ entanglement
            qml.CZ(wires=[0, 1])
            # Second RZ layer
            for i in range(2):
                qml.RZ(x[i], wires=i)

        return qml.state()

    start_zz = time.time()

    # Pre-compute states
    print("   Pre-computing quantum states...")
    states_train_zz = [get_state_zz(x) for x in X_train_scaled]
    states_test_zz = [get_state_zz(x) for x in X_test_scaled]

    # Compute kernel matrix
    print("   Computing kernel matrices...")
    K_train_zz = np.zeros((500, 500))
    for i in range(500):
        for j in range(i, 500):
            val = qml.math.fidelity_statevector(states_train_zz[i], states_train_zz[j])
            K_train_zz[i,j] = val
            K_train_zz[j,i] = val

    K_test_zz = np.zeros((250, 500))
    for i in range(250):
        for j in range(500):
            K_test_zz[i,j] = qml.math.fidelity_statevector(states_test_zz[i], states_train_zz[j])

    # Train QSVM
    qsvm_zz = SVC(kernel='precomputed', probability=True)
    qsvm_zz.fit(K_train_zz, y_train_500)
    y_pred_zz = qsvm_zz.predict(K_test_zz)
    y_score_zz = qsvm_zz.decision_function(K_test_zz)
    y_score_norm_zz = (y_score_zz - y_score_zz.min()) / (y_score_zz.max() - y_score_zz.min() + 1e-10)

    time_zz = time.time() - start_zz

    # Store results
    results_zz.append({
        'run': run+1,
        'circuit': 'ZZFeatureMap (WITH ZZ)',
        'accuracy': accuracy_score(y_test_250, y_pred_zz) * 100,
        'precision': precision_score(y_test_250, y_pred_zz, average='binary'),
        'recall': recall_score(y_test_250, y_pred_zz, average='binary'),
        'f1': f1_score(y_test_250, y_pred_zz, average='binary'),
        'roc_auc': roc_auc_score(y_test_250, y_score_norm_zz),
        'time': time_zz
    })

    print(f"   ✅ Acc={results_zz[-1]['accuracy']:.1f}%, F1={results_zz[-1]['f1']:.3f}, Time={time_zz:.1f}s")

# ============================================
# RESULTS
# ============================================
print("\n" + "="*60)
print("📊 ZZFEATUREMAP (WITH ZZ) - RESULTS")
print("="*60)

df_zz = pd.DataFrame(results_zz)

print(f"\n{'Metric':<15} {'Mean':<10} {'Std':<10} {'95% CI':<20}")
print("-" * 55)

for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc', 'time']:
    values = df_zz[metric].values
    mean_val = np.mean(values)
    std_val = np.std(values)
    ci = stats.t.interval(0.95, len(values)-1, loc=mean_val, scale=stats.sem(values))

    if metric == 'accuracy':
        print(f"{metric.capitalize():<15} {mean_val:.1f}%    ±{std_val:.1f}%    [{ci[0]:.1f}%, {ci[1]:.1f}%]")
    elif metric == 'time':
        print(f"{metric.capitalize():<15} {mean_val:.1f}s   ±{std_val:.1f}s   [{ci[0]:.1f}s, {ci[1]:.1f}s]")
    else:
        print(f"{metric.capitalize():<15} {mean_val:.3f}    ±{std_val:.3f}    [{ci[0]:.3f}, {ci[1]:.3f}]")

# Save
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"pennylane_zz_WITHzz_500_{timestamp}.csv"
df_zz.to_csv(filename, index=False)
print(f"\n💾 Saved to {filename}")
print("\n✅ CELL 2 COMPLETE!")
```

```python
# ============================================
# CELL 4: CIRQ CONFIGURATION A (Basic H+RY, NO ZZ)
# 500 samples, 7 runs
# ============================================

!pip install cirq scikit-learn pandas --quiet

import numpy as np
import cirq
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import time
import pandas as pd
from scipy import stats
from datetime import datetime

print("="*60)
print("🔹 CELL 4: CIRQ CONFIGURATION A (Basic H+RY, NO ZZ)")
print("500 samples, 7 runs")
print("="*60)

# Load MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target.astype(int)
mask = (y == 0) | (y == 1)
X, y = X[mask], y[mask]

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store results
results_cirq_basic = []

for run in range(7):
    print(f"\n🔷 Run {run+1}/7 (seed={42+run})")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.3, random_state=42+run, stratify=y
    )

    # PCA to 2D
    pca = PCA(n_components=2)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # TAKE EXACTLY 500 SAMPLES
    X_train_500 = X_train_pca[:500]
    y_train_500 = y_train[:500]
    X_test_250 = X_test_pca[:250]
    y_test_250 = y_test[:250]

    # Scale for quantum circuits
    X_train_scaled = X_train_500 * np.pi
    X_test_scaled = X_test_250 * np.pi

    print(f"   Train: {len(X_train_500)} samples, Test: {len(X_test_250)} samples")

    # BASIC CIRCUIT - NO ZZ
    def create_basic_circuit(x, qubits):
        """Basic H + RY encoding - NO ZZ"""
        circuit = cirq.Circuit()
        for i, qubit in enumerate(qubits):
            circuit.append(cirq.H(qubit))
            circuit.append(cirq.ry(x[i])(qubit))
        return circuit

    def cirq_kernel_basic(x1, x2, n_qubits):
        """Compute kernel value for basic circuit"""
        qubits = [cirq.GridQubit(0, i) for i in range(n_qubits)]
        circuit1 = create_basic_circuit(x1, qubits)
        circuit2 = create_basic_circuit(x2, qubits)

        full_circuit = cirq.Circuit()
        full_circuit.append(cirq.inverse(circuit1))
        full_circuit.append(circuit2)

        simulator = cirq.Simulator()
        result = simulator.simulate(full_circuit)
        return np.abs(result.final_state_vector[0]) ** 2

    start_time = time.time()

    # Compute kernel matrices
    print("   Computing kernel matrices...")
    K_train = np.zeros((500, 500))
    for i in range(500):
        for j in range(i, 500):
            val = cirq_kernel_basic(X_train_scaled[i], X_train_scaled[j], 2)
            K_train[i,j] = val
            K_train[j,i] = val

    K_test = np.zeros((250, 500))
    for i in range(250):
        for j in range(500):
            K_test[i,j] = cirq_kernel_basic(X_test_scaled[i], X_train_scaled[j], 2)

    # Train QSVM
    qsvm = SVC(kernel='precomputed', probability=True)
    qsvm.fit(K_train, y_train_500)
    y_pred = qsvm.predict(K_test)
    y_score = qsvm.decision_function(K_test)
    y_score_norm = (y_score - y_score.min()) / (y_score.max() - y_score.min() + 1e-10)

    elapsed_time = time.time() - start_time

    # Store results
    results_cirq_basic.append({
        'run': run+1,
        'framework': 'Cirq',
        'circuit': 'Basic H+RY (NO ZZ)',
        'accuracy': accuracy_score(y_test_250, y_pred) * 100,
        'precision': precision_score(y_test_250, y_pred, average='binary'),
        'recall': recall_score(y_test_250, y_pred, average='binary'),
        'f1': f1_score(y_test_250, y_pred, average='binary'),
        'roc_auc': roc_auc_score(y_test_250, y_score_norm),
        'time': elapsed_time
    })

    print(f"   ✅ Acc={results_cirq_basic[-1]['accuracy']:.1f}%, F1={results_cirq_basic[-1]['f1']:.3f}, Time={elapsed_time:.1f}s")

# Save results
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
df = pd.DataFrame(results_cirq_basic)
filename = f"cirq_basic_NOzz_500_{timestamp}.csv"
df.to_csv(filename, index=False)
print(f"\n💾 Saved to {filename}")
print("\n✅ CELL 4 COMPLETE! Run CELL 5 next.")
```

###Configuration B

```python
# ============================================
# CELL 5: CIRQ CONFIGURATION B (ZZFeatureMap, WITH ZZ)
# 500 samples, 7 runs
# ============================================

!pip install cirq scikit-learn pandas --quiet

import numpy as np
import cirq
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import time
import pandas as pd
from scipy import stats
from datetime import datetime

print("="*60)
print("🔸 CELL 5: CIRQ CONFIGURATION B (ZZFeatureMap, WITH ZZ)")
print("500 samples, 7 runs")
print("="*60)

# Load MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target.astype(int)
mask = (y == 0) | (y == 1)
X, y = X[mask], y[mask]

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Store results
results_cirq_zz = []

for run in range(7):
    print(f"\n🔷 Run {run+1}/7 (seed={42+run})")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.3, random_state=42+run, stratify=y
    )

    # PCA to 2D
    pca = PCA(n_components=2)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # TAKE EXACTLY 500 SAMPLES
    X_train_500 = X_train_pca[:500]
    y_train_500 = y_train[:500]
    X_test_250 = X_test_pca[:250]
    y_test_250 = y_test[:250]

    # Scale for quantum circuits
    X_train_scaled = X_train_500 * np.pi
    X_test_scaled = X_test_250 * np.pi

    print(f"   Train: {len(X_train_500)} samples, Test: {len(X_test_250)} samples")

    # ZZFEATUREMAP CIRCUIT - WITH ZZ
    def create_zz_circuit(x, qubits):
        """ZZFeatureMap with 2 repetitions"""
        circuit = cirq.Circuit()

        # Initial Hadamards
        for qubit in qubits:
            circuit.append(cirq.H(qubit))

        # 2 repetitions
        for _ in range(2):
            # RZ rotations
            for i, qubit in enumerate(qubits):
                circuit.append(cirq.rz(x[i])(qubit))

            # CZ entanglement
            for i in range(len(qubits)-1):
                circuit.append(cirq.CZ(qubits[i], qubits[i+1]))

            # Second RZ layer
            for i, qubit in enumerate(qubits):
                circuit.append(cirq.rz(x[i])(qubit))

        return circuit

    def cirq_kernel_zz(x1, x2, n_qubits):
        """Compute kernel value for ZZFeatureMap"""
        qubits = [cirq.GridQubit(0, i) for i in range(n_qubits)]
        circuit1 = create_zz_circuit(x1, qubits)
        circuit2 = create_zz_circuit(x2, qubits)

        full_circuit = cirq.Circuit()
        full_circuit.append(cirq.inverse(circuit1))
        full_circuit.append(circuit2)

        simulator = cirq.Simulator()
        result = simulator.simulate(full_circuit)
        return np.abs(result.final_state_vector[0]) ** 2

    start_time = time.time()

    # Compute kernel matrices
    print("   Computing kernel matrices...")
    K_train = np.zeros((500, 500))
    for i in range(500):
        for j in range(i, 500):
            val = cirq_kernel_zz(X_train_scaled[i], X_train_scaled[j], 2)
            K_train[i,j] = val
            K_train[j,i] = val

    K_test = np.zeros((250, 500))
    for i in range(250):
        for j in range(500):
            K_test[i,j] = cirq_kernel_zz(X_test_scaled[i], X_train_scaled[j], 2)

    # Train QSVM
    qsvm = SVC(kernel='precomputed', probability=True)
    qsvm.fit(K_train, y_train_500)
    y_pred = qsvm.predict(K_test)
    y_score = qsvm.decision_function(K_test)
    y_score_norm = (y_score - y_score.min()) / (y_score.max() - y_score.min() + 1e-10)

    elapsed_time = time.time() - start_time

    # Store results
    results_cirq_zz.append({
        'run': run+1,
        'framework': 'Cirq',
        'circuit': 'ZZFeatureMap (WITH ZZ)',
        'accuracy': accuracy_score(y_test_250, y_pred) * 100,
        'precision': precision_score(y_test_250, y_pred, average='binary'),
        'recall': recall_score(y_test_250, y_pred, average='binary'),
        'f1': f1_score(y_test_250, y_pred, average='binary'),
        'roc_auc': roc_auc_score(y_test_250, y_score_norm),
        'time': elapsed_time
    })

    print(f"   ✅ Acc={results_cirq_zz[-1]['accuracy']:.1f}%, F1={results_cirq_zz[-1]['f1']:.3f}, Time={elapsed_time:.1f}s")

# Save results
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
df = pd.DataFrame(results_cirq_zz)
filename = f"cirq_zz_WITHzz_500_{timestamp}.csv"
df.to_csv(filename, index=False)
print(f"\n💾 Saved to {filename}")
print("\n✅ CELL 5 COMPLETE! Run CELL 6 for graphs.")
```
